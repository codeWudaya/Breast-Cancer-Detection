# Breast Cancer Detection using ML By codeXudaya

# Import libraries

```python
import numpy as np
import pandas as pd
```

# Load the Breast Cancer Wisconsin dataset

```python
#load
from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
```

check the data

```python
# target_names
df.target_names
```

```python
# target
df.target
```

```python
# 0/1 
check={i : target  for i,target in enumerate(df.target_names)}
check
```

```python
# feature_names
df.feature_names
```

```python
# data
df.data
```

# Create a DataFrame 

```python
data=pd.DataFrame(df.data,columns=df.feature_names)
data
```

```python
#add label
data['lebal']=df.target
```

```python
data
```

# features and target variable

```python
#features
X=data.drop(columns='lebal')
X
```

```python
# target
y=data['lebal']
y
```

# train test split

```python
# import
from sklearn.model_selection import train_test_split
```

```python
#,,,
x_train,x_test,y_train,y_test=train_test_split(X , y ,test_size=0.2 , random_state=2)
```

```python
X.shape,y.shape
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Model fit

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

# check accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=model.predict(x_test)
```

```python
# y_pred and y_test
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy.round(2)
```

# Prediction on new data points

```python
new_data = np.array([13.54, 14.36, 87.46, 566.3, 0.09779, 0.08129, 0.06664, 0.04781, 0.1885, 0.05766,
                     0.2699, 0.7886, 2.058, 23.56, 0.008462, 0.0146, 0.02387, 0.01315, 0.0198, 0.0023,
                     15.11, 19.26, 99.7, 711.2, 0.144, 0.1773, 0.239, 0.1288, 0.2977, 0.07259]).reshape(1, -1)

prediction = model.predict(new_data)
result = 'Benign' if prediction[0] == 1 else 'Malignant'
print(f'The Breast Cancer is {result}')
```
